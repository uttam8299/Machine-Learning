# Question 1

```python
# Program for taking one number a input and then using if,elif abd else concept,
# assign a grade to the input number from
# A,B,C,D,E,Fail
```

```python
num = int(input("Enter the number"))

if num in range(80,100):
    print("|-----|")            
    print("|  A  |")
    print("|-----|")
elif num in range(60,80):
    print("|-----|")
    print("|  B  |")
    print("|-----|")
elif num in range(50,60):
    print("|-----|")
    print("|  C  |")
    print("|-----|")
elif num in range(40,50):
    print("|-----|")
    print("|  D  |")
    print("|-----|")
elif num in range(33,40):
    print("|-----|")
    print("|  E  |")
    print("|-----|")
else:
    print("|-----|")
    print("|Fail |")
    print("|-----|")

    
```

# Question 2

```python
# random number from system and guess from player
```

```python
import random

guess = int(random.randrange(1,251))
print("Number generated by computer : " )
print(guess)
ply_guess = int(input("Enter your guess number : "))
 
    
if(ply_guess<1 or ply_guess>250):
    print("Reduce your expectation for 20-20 Cricket")
if(abs(ply_guess-guess)<10):
    print("Close By, you are True Indian Fan!")
else:
    print("You don't watch that much! :P")
```
